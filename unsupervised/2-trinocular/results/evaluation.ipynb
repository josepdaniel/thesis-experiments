# Trinocular, Grayscale

```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_sequence(sequence):

    pose_predictions = np.load(f"./{sequence}/poses.npy")
    actual_poses = np.load(f"/home/joseph/Documents/epidata/module-1-1/{sequence}/poses_gt_relative.npy")
    actual_poses_absolute = np.load(f"/home/joseph/Documents/epidata/module-1-1/{sequence}/poses_gt_absolute.npy")
    print(pose_predictions.shape)
    print(actual_poses.shape)

    xs_pred = -1*np.cumsum(pose_predictions[:, 0]) * 3
    ys_pred = np.cumsum(pose_predictions[:, 1]) * 3
    zs_pred = np.cumsum(pose_predictions[:, 2]) * 4

    xs_actual = np.cumsum(actual_poses[:, 0])
    ys_actual = np.cumsum(actual_poses[:, 1])
    zs_actual = np.cumsum(actual_poses[:, 2])


    xs_actual_absolute = np.cumsum(actual_poses_absolute[:, 0])
    ys_actual_absolute = np.cumsum(actual_poses_absolute[:, 1])
    zs_actual_absolute = np.cumsum(actual_poses_absolute[:, 2])

    %matplotlib inline

    fig = plt.figure(figsize = (15,5))
    ax1 = fig.add_subplot(121)
    ax1.plot(xs_pred, zs_pred*-1)
    ax1.plot(xs_actual, zs_actual, c='r')
    ax1.set_title("Predicted vs Actual Trajectory (XZ plane)")
    ax1.axis('equal')

    
    ax2 = fig.add_subplot(122)
    ax2.plot(xs_pred, ys_pred)
    ax2.plot(xs_actual, ys_actual, c='r')
    ax2.set_title("Ground Truth Trajectory (XY plane)")
    ax2.axis('equal')


    fig.suptitle(f'{sequence}', fontsize=16)
```

```python
plot_sequence("seq4")
```

```python
plot_sequence("seq2")
```
